```python
connection=psycopg2.connect(
                            host="localhost",
                            port="xxxx",
                            database="python_db",
                            user="postgres",
                            password="xxx"
                            )

def get_hospital_detail(h_id):
    con=psycopg2.connect(
                            host="localhost",
                            port="xxxx",
                            database="python_db",
                            user="postgres",
                            password="xxx"
                            )
    cursor=con.cursor()
    query="""SELECT * FROM hospital where hospital_id=(%s)"""
    idx=str(h_id)
    cursor.execute(query, idx)
    record=cursor.fetchall()
    connection.close()
    return record

x=get_hospital_detail(1)
print(''.join('Printing Hospital record\n Hospital_id: {}\n Hospital Name: {}\n Bed count: {}'.format(*p) for p in x))
```

```python
def get_specialist_doctors_list(speciality, salary):
    con=psycopg2.connect(
                            host="localhost",
                            port="xxxx",
                            database="python_db",
                            user="postgres",
                            password="xxx"
                            )
    cursor=con.cursor()
    query="""SELECT * from doctor where speciality=(%s) and salary>(%s)"""
    values=(speciality, str(salary))
    cursor.execute(query, values)
    record=cursor.fetchall()
    connection.close()
    return record

y=get_specialist_doctors_list("Garnacologist", 30000)
print("".join("\n\nPrinting doctors whose specialty is Garnacologist and salary greater than 30000\n Doctor Id: {}\n Doctor Name: {}\n Hospital Id: {}\n Joining Date: {}\n Specialty: {}\n Salary: {}\n Experience:{}".format(*p) for p in y))
```

```python
def get_doctors(hospital_id):
    con=psycopg2.connect(
                            host="localhost",
                            port="xxxx",
                            database="python_db",
                            user="postgres",
                            password="xxx"
                            )
    cursor=con.cursor()
    query="""SELECT d.doctor_name, h.hospital_name
             FROM doctor d
             join hospital h
             on d.hospital_id=h.hospital_id
             where h.hospital_id=%s
                """
    cursor.execute(query,str(hospital_id))
    record=cursor.fetchall()
    con.close()
    return record

get_doctors(2)
for row in get_doctors(2):
    print("Doctors from the hospital {}: \n Name: {}\n ".format(row[1], row[0]))
```

```python
import pandas as pd


def years_exp(doctor_id):
    con=psycopg2.connect(
                            host="localhost",
                            port="xxxx",
                            database="python_db",
                            user="postgres",
                            password="xxx"
                            )
    cursor=con.cursor()
    query="""SELECT DATE_PART('year', CURRENT_DATE) - DATE_PART('year', joining_date) from doctor where doctor_id=(%s)"""
#     query="SELECT * from doctor where doctor_id=%s"
    cursor.execute(query, (doctor_id,))
    record=cursor.fetchall()
    con.close()
    return record[0]

# years_exp(102)
    

def update_doctor_experience(doctor_id):
    con=psycopg2.connect(
                            host="localhost",
                            port="xxxx",
                            database="python_db",
                            user="postgres",
                            password="xxx"
                            )
    cursor=con.cursor()
    query="""Update doctor set experience=(%s) where doctor_id=%s"""
    cursor.execute(query, (years_exp(doctor_id), doctor_id))
    con.commit()
    cursor.execute("Select * from doctor")
    record=pd.DataFrame(cursor.fetchall())
    con.close()
    return record

update_doctor_experience(102)
```
